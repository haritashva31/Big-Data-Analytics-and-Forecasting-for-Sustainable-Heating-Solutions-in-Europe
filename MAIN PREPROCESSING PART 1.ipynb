```python
from pyspark.sql import SparkSession

# Create a Spark session
spark = SparkSession.builder.appName("When2Heat").getOrCreate()
```

```python
# File path
file_path = "/Downloads/opsd-when2heat-2023-07-27/when2heat.csv"

# Read the CSV file into a DataFrame
df = spark.read.option("delimiter", ";").csv(file_path, header=True)
```

```python
# Check the number of columns
num_columns = len(df.columns)
print(f"Number of columns: {num_columns}")

# Check the column names
column_names = df.columns
print("Column names:", column_names)
```

```python
df.printSchema()
```

```python
# Drop 'utc_timestamp' column
df = df.drop('utc_timestamp')

# Check the updated schema
df.printSchema()
```

```python
from pyspark.sql.functions import to_timestamp
from pyspark.sql.types import TimestampType
df = df.withColumn('cet_cest_timestamp', to_timestamp(df['cet_cest_timestamp']).cast(TimestampType()))
```

```python
generic_column_names = ['COP_ASHP_floor', 'COP_ASHP_radiator', 'COP_ASHP_water',
                        'COP_GSHP_floor', 'COP_GSHP_radiator', 'COP_GSHP_water',
                        'COP_WSHP_floor', 'COP_WSHP_radiator', 'COP_WSHP_water',
                        'heat_demand_space', 'heat_demand_space_COM', 'heat_demand_space_MFH',
                        'heat_demand_space_SFH', 'heat_demand_total', 'heat_demand_water',
                        'heat_demand_water_COM', 'heat_demand_water_MFH', 'heat_demand_water_SFH',
                        'heat_profile_space_COM', 'heat_profile_space_MFH', 'heat_profile_space_SFH',
                        'heat_profile_water_COM', 'heat_profile_water_MFH', 'heat_profile_water_SFH']

country_codes = ['SE', 'EE', 'NO', 'DK', 'GR', 'IE', 'SK', 'PT', 'BE', 'SI', 'LU', 'IT', 'NL', 'HR', 'ES', 'CZ', 'AT', 'DE', 'HU', 'RO', 'LT', 'FR', 'BG', 'CH', 'FI', 'GB', 'LV', 'PL']
```

```python
from pyspark.sql.functions import col
from pyspark.sql.types import FloatType

# List of columns to convert to float
columns_to_convert = df.columns[1:]  # Exclude the first timestamp column

# Convert string columns to float
for column in columns_to_convert:
    df = df.withColumn(column, col(column).cast(FloatType()))

# Check the updated schema
df.printSchema()
```

```python
from pyspark.sql.functions import year, month, dayofmonth, hour
df = df.withColumn("year", year("cet_cest_timestamp"))
df = df.withColumn("month", month("cet_cest_timestamp"))
df = df.withColumn("day", dayofmonth("cet_cest_timestamp"))
df = df.withColumn("hour", hour("cet_cest_timestamp"))
```

```python
from pyspark.sql.functions import col

# Assuming df is your DataFrame with columns like 'AT_COP_ASHP_floor', 'BE_COP_ASHP_floor', etc.
# Assuming 'cet_cest_timestamp' is your timestamp column

# Extract country code from column names
country_codes = ['SE', 'EE', 'NO', 'DK', 'GR', 'IE', 'SK', 'PT', 'BE', 'SI', 'LU', 'IT', 'NL', 'HR', 'ES', 'CZ', 'AT', 'DE', 'HU', 'RO', 'LT', 'FR', 'BG', 'CH', 'FI', 'GB', 'LV', 'PL']

# Create country-wise DataFrames
country_dfs = {}

for code in country_codes:
    # Select columns for the current country
    selected_columns = ['cet_cest_timestamp', 'year', 'month', 'day', 'hour'] + [col_name for col_name in df.columns if col_name.startswith(code)]
    country_df = df.select(selected_columns)
    
    # Rename columns for consistency
    rename_mapping = {col_name: col_name[len(code) + 1:] for col_name in selected_columns[5:]}
    country_df = country_df.select(['cet_cest_timestamp', 'year', 'month', 'day', 'hour'] + [col(col_name).alias(rename_mapping[col_name]) for col_name in country_df.columns[5:]])
    
    # Store the country DataFrame
    country_dfs[code] = country_df
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Iterate through country DataFrames
for country_code, country_df in country_dfs.items():
    print(f"Processing {country_code} DataFrame...")

    # Iterate through columns containing 'COP' from the generic list
    for column in [col for col in generic_column_names if 'COP' in col]:
        print(f"Processing column: {column}")

        # Define a window specification for calculating the mean
        window_spec = Window().partitionBy('year', 'month')

        # Replace null values with the daily mean of that specific month and year
        country_df = country_df.withColumn(column, F.coalesce(country_df[column], F.mean(column).over(window_spec)))

    # Update the country DataFrame in the dictionary
    country_dfs[country_code] = country_df
```

```python
from pyspark.sql import functions as F
from pyspark.sql.window import Window

# Iterate through country DataFrames
for country_code, country_df in country_dfs.items():
    print(f"Processing {country_code} DataFrame...")

    # Check if the country DataFrame contains heat_demand columns
    heat_demand_columns = [col for col in generic_column_names if 'heat_demand' in col]
    if not set(heat_demand_columns).issubset(country_df.columns):
        print(f"Skipping {country_code} DataFrame as it doesn't contain heat_demand columns.")
        continue

    # Iterate through heat_demand columns
    for column in heat_demand_columns:
        print(f"Processing column: {column}")

        # Define a window specification for linear interpolation
        window_spec = Window().partitionBy('year', 'month').orderBy('day')

        # Replace 0.0 values with linear interpolation based on surrounding data
        country_df = country_df.withColumn(
            column,
            F.when(country_df[column] == 0.0, F.coalesce(
                F.first(column, ignorenulls=True).over(window_spec),
                F.last(column, ignorenulls=True).over(window_spec),
                F.expr(f"LAG({column}, 1) OVER (ORDER BY day)"),
                F.expr(f"LEAD({column}, 1) OVER (ORDER BY day)")
            )).otherwise(country_df[column])
        )

    # Update the country DataFrame in the dictionary
    country_dfs[country_code] = country_df
```
